```python
#Rastgelelik için random kütüphanesinin yüklenmesi
import random

#Oyun fonksiyonun tanımlanması
def tas_kagit_makas_ALİ_NİHAT_ÖĞÜT():
    # Oyun kurallarının oyuncuya bildirilmesi
    print("Kurallar:")
    print("1. Taş, makası yener.")
    print("2. Kağıt, taşı yener.")
    print("3. Makas, kağıdı yener.")
    print("İki tur kazanan oyuncu oyunun galibi olur.")
    print("Taş seçmek için 't',Kağıt seçmek için 'k', Makas seçmek için 'm' yazın.")
    print("Başarılar!")

    # Genel oyun döngüsü
    while True:
        oyuncu_kazandi = 0
        bilgisayar_kazandi =0
        #Oyuncu veya bilgisayar 2 skora alana kadar çalışacak döngü
        while oyuncu_kazandi < 2 and bilgisayar_kazandi < 2:
            oyuncu_secimi = None
            #Oyuncu geçerli bir seçim yapana kadar çalışacak döngü(Bu döngünün içinde
             #T K ve M olacak şekilde seçenekler büyük harf ile girilmiş olsa dahi çalışmasını sağlayacak .lower metodu bulunmakta.)
            while oyuncu_secimi not in ['t', 'k', 'm']:
                oyuncu_secimi = input("Seçiminizi yapın ('t' için Taş, 'k' için Kağıt, 'm' için Makas): ").lower()

            secenekler = {'t': 'taş', 'k': 'kağıt', 'm': 'makas'}
            oyuncu_secimi_str = secenekler[oyuncu_secimi]
            bilgisayar_secimi = random.choice(list(secenekler.values()))

            print(f"Bilgisayarın seçimi: {bilgisayar_secimi}")
            #Beraberlik durumu
            if oyuncu_secimi_str == bilgisayar_secimi:
                sonuc = "Beraberlik"
            #Oyuncunun kazanma durumları
            elif (oyuncu_secimi_str == "taş" and bilgisayar_secimi == "makas") or \
                 (oyuncu_secimi_str == "kağıt" and bilgisayar_secimi == "taş") or \
                 (oyuncu_secimi_str == "makas" and bilgisayar_secimi == "kağıt"):
                sonuc = "Oyuncu Kazandı"
                oyuncu_kazandi += 1
            #Bilgisayarın kazanma durumu
            else:
                sonuc = "Bilgisayar Kazandı"
                bilgisayar_kazandi += 1
            #Tur sonucunun ve skor tablosunun yazdırılması
            print(f"Sonuç: {sonuc}")
            print(f"Mevcut Skor - Oyuncu: {oyuncu_kazandi} Bilgisayar: {bilgisayar_kazandi}")

        # Oyun galibinin belirlenmesi ve yazdırılması
        if oyuncu_kazandi > bilgisayar_kazandi:
            print("Oyun Bitti! Oyuncu Kazandı!")
        else:
            print("Oyun Bitti! Bilgisayar Kazandı!")

        # Tekrar oyun oynama isteği
        tekrar_oyna = input("Başka bir oyun oynamak ister misiniz? (E/H): ").strip().lower()
        #Bilgisayar için rastgele olacak şekilde tekrar oynama isteği
        bilgisayar_oyna = random.choice(['e', 'h'])
        print(f"Bilgisayarın cevabı: {'Evet' if bilgisayar_oyna == 'e' else 'Hayır'}")
        # Her iki taraf da oynamak isterse oyun devam eder
        if tekrar_oyna == 'e' and bilgisayar_oyna == 'e':
            print("Her iki taraf da oyuna devam etmek istiyor. Yeni oyun başlıyor!")
            continue
        else:
            if tekrar_oyna != 'e':
                print("Oyuncu oyunu sonlandırmak istedi.")
            if bilgisayar_oyna != 'e':
                print("Bilgisayar oyunu sonlandırmak istedi.")
            print("İyi oyundu, tebrikler")
            break


# Fonksiyonu çalıştırma
if __name__ == "__main__":
    tas_kagit_makas_ALİ_NİHAT_ÖĞÜT()
```
